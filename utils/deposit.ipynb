```python

import numpy as np
import matplotlib.pyplot as plt 
import math 
```

```python
x1 = []
y1 = []

power = 25
time = 0
harvested = 0

while(time < 100000):
    harvested += power
    time += math.ceil(0.001 * (harvested ** 1.2));
    x1.append(time)
    y1.append(harvested / time)
```

```python
plt.plot(x1, y1)
plt.show()
```

```python
z1 = []
for i in range(len(x1)):
    z1.append(y1[i] / (x1[i] + 50000))

plt.plot(x1, z1)
plt.show()
```

```python
def max_harv_rate(power):
    time = 0
    harvested = 0
    maxrate = 0

    while(time < 100000):
        harvested += power
        time += math.ceil(0.001 * (harvested ** 1.2))
        rate = harvested / (time + 50000)
        maxrate = max(maxrate, rate)
        if (rate < maxrate):
            break

    return maxrate
```

```python
x = []
y = []

for i in range(1, 201):
    x.append(i)
    y.append(max_harv_rate(i))

plt.plot(x, y)
plt.show()
```
